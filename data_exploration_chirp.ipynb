```python
keys = list(dat['spiketimes'].keys())
spiketimes = dat['spiketimes'] # spike times in frames

sampling_frequency = 20000.0
```

# load events

```python
fp = f'/user/sebert/home/Documents/Experiments/Spatiotemporal_tuning_curves/analisis_chirp_led/event_list_LED_.csv'
events = pd.read_csv(fp)
events

# start_event  and end_event contain the start  and end TIMES for each trial, in sampling frequency (?)
```

```python
# group by intenisty
g = events.groupby('extra_description')
intensities = list(g.groups.keys())
intensities
```

# align spikes and stimulus in each trial

```python
# exract raster per trial and calculate PSTH

nb_trials = 10

dt = 0.06*sampling_frequency
stim_length = events['event_duration'].values[0]
time = np.arange(0,stim_length,dt)
bin_edges = np.append(time, 2 * time[-1] - time[-2])
sigma = 1


data = {}

for i in intensities:
    data[i] = {}
    evts = g.get_group(i)
    start = evts['start_event'].values
    end = evts['end_event'].values

    data[i]['stimuli'] = []
    for nb in range(nb_trials):
        data[i]['stimuli'].append(stim[start[nb]:end[nb]+1])

    for key in keys: 
        data[i][key] = {}
        spktms = spiketimes[key][()].flatten()
        df = pd.DataFrame(spktms,  columns = ['Times'])
        rasters = []
        for nb in range(nb_trials):
            times = df.query(f"Times>={start[nb]} & Times<={end[nb]}") - start[nb]
            rasters.append(np.asarray(times["Times"]))


        spike_times_all_trials = np.concatenate(rasters, axis= 0)
        count, bins = np.histogram(spike_times_all_trials, bins=bin_edges, density=False) 
        #count_smooth = gaussian_filter(count,sigma)

        data[i][key]['rasters'] = rasters
        data[i][key]['count'] = count

```

# solve stimulus alignment

```python
data_old = data
```

```python
# detect the start of the flash in the stimulus
flashstarts_i = []
for i in intensities: 
   flashstarts = []

   fig, ax = plt.subplots(3,1,sharex = True)
   for s in data_old[i]['stimuli']:
      ax[0].plot(s)
      ax[1].plot(np.diff(s))
      ax[2].plot(np.diff(s)>250)

      dets = np.diff(s)>250
      xf = np.where(dets ==True)
      flash_start = xf[0][0] 
      ax[0].axvline(flash_start)
      flashstarts.append(flash_start)
   flashstarts_i.append(flashstarts)
```

```python
# exract raster per trial and calculate PSTH
# align event start and end to start of the first flash in the stimulus

nb_trials = 10

dt = 0.06*sampling_frequency
stim_length = events['event_duration'].values[0]
time = np.arange(0,stim_length,dt)

stim_length_s = stim_length/sampling_frequency
dt_s= dt/sampling_frequency
time_s = np.arange(0,stim_length_s,dt_s)
stime_s = np.arange(0,stim_length_s,1/sampling_frequency)
bin_edges = np.append(time_s, 2 * time_s[-1] - time_s[-2])


sigma = 1


data ={}
for x,i in enumerate(intensities):
    data[i] = {}
    evts = g.get_group(i)
    start = evts['start_event'].values + flashstarts_i[x] -1000
    end = evts['end_event'].values + flashstarts_i[x] - 100


    data[i]['stimuli'] = []
    for nb in range(nb_trials):
        stimm = stim[start[nb]:end[nb]+1]
        data[i]['stimuli'].append(stimm[:int(stim_length)])
        #data[i]['stimuli'].append(stimm)

    for key in keys: 
        data[i][key] = {}
        spktms = spiketimes[key][()].flatten()
        df = pd.DataFrame(spktms,  columns = ['Times'])
        rasters = []
        for nb in range(nb_trials):
            times = df.query(f"Times>={start[nb]} & Times<={end[nb]}") - start[nb]
            times = times.div(sampling_frequency)
            rasters.append(np.asarray(times["Times"]))


        spike_times_all_trials = np.concatenate(rasters, axis= 0)
        count, bins = np.histogram(spike_times_all_trials, bins=bin_edges, density=False) 
        count_smooth = gaussian_filter(count,sigma)/nb_trials

        data[i][key]['rasters'] = rasters
        data[i][key]['count'] = count
```
